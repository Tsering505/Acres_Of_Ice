```python
file_path = 'C:\\Users\\lenovo\\Downloads\\AOI_Data(Site_Wise)\\Likir\\Likir_data_11-29.csv'
df_11=pd.read_csv(file_path)
df_11.columns=df_11.columns.str.strip()
df_11.rename(columns={'Likir Time':'Time'},inplace=True)


file_path = 'C:\\Users\\lenovo\\Downloads\\AOI_Data(Site_Wise)\\final_Likir.csv'
df_11.to_csv(file_path,index=False)
df_Likir=df_11
df_Likir['site_name']='Likir'
```

```python
file_path='C:\\Users\\lenovo\\Downloads\\AOI_Data(Site_Wise)\\Ursi\\Nov 28\\DATA.csv'
df_12=pd.read_csv(file_path,on_bad_lines='warn')
df_12.columns=df_12.columns.str.strip()

file_path='C:\\Users\\lenovo\\Downloads\\AOI_Data(Site_Wise)\\Ursi\\Ursi_data_11-16.csv'
df_13=pd.read_csv(file_path,on_bad_lines='warn')
df_13.columns=df_13.columns.str.strip()

df_Ursi=pd.concat([df_12,df_13],ignore_index=True)
df_Ursi['site_name']='Ursi'
```

```python
df_Igoo['Time'] = pd.to_datetime(df_Igoo['Time'], errors='coerce')  # coerce invalid parsing to NaT
df_Igoo['Temp'] = pd.to_numeric(df_Igoo['Temp'], errors='coerce')  # Coerce invalid values to NaN
df_Ursi['Time'] = pd.to_datetime(df_Ursi['Time'], errors='coerce')  # coerce invalid parsing to NaT
df_Ursi['Temp'] = pd.to_numeric(df_Ursi['Temp'], errors='coerce')  # Coerce invalid values to NaN
df_Ayee['Time'] = pd.to_datetime(df_Ayee['Time'], errors='coerce')  # coerce invalid parsing to NaT
df_Ayee['Temp'] = pd.to_numeric(df_Ayee['Temp'], errors='coerce')  # Coerce invalid values to NaN
df_Likir['Time'] = pd.to_datetime(df_Likir['Time'], errors='coerce')  # coerce invalid parsing to NaT
df_Likir['Temp'] = pd.to_numeric(df_Likir['Temp'], errors='coerce')  # Coerce invalid values to NaN
```

```python
df_Igoo=df_Igoo.sort_values(by='Time')
df_Ursi=df_Ursi.sort_values(by='Time')
df_Likir=df_Likir.sort_values(by='Time')
df_Ayee=df_Ayee.sort_values(by='Time')
```

```python

plt.figure(figsize=(10, 6))  
plt.plot(df_Igoo['Time'], df_Igoo['Temp'], marker='.', color='b', label='Igoo',linestyle='-')
plt.plot(df_Ursi['Time'], df_Ursi['Temp'], marker='.', color='g', label='Ursi',linestyle='--')
plt.plot(df_Ayee['Time'], df_Ayee['Temp'], marker='.', color='r', label='Ayee',linestyle=':')
plt.plot(df_Likir['Time'], df_Likir['Temp'], marker='.', color='c', label='Likir',linestyle='-')
plt.legend(title='Site Name', loc='best')
plt.xticks(rotation=45)
plt.show()
```
